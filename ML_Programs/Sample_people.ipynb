```python
data = {
    'Name': ['Amit', 'Priya', 'Rohan', 'Sneha', None],
    'Age': [23, 25, None, 22, 24],
    'City': ['Delhi', None, 'Mumbai', 'Pune', 'Chennai'],
    'Marks': [85, 90, 78, None, 88]
}
df = pd.DataFrame(data)
df_fillna = df.fillna({
    'Name': 'Unknown',
    'Age': df['Age'].mean(),
    'City': 'Not Specified',
    'Marks': df['Marks'].mean()
    })
print(df_fillna)
```
