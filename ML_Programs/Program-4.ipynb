 # Program-4: Use of arange, linspace, rand, randint, randn, normal.


```python
import numpy as np
import random
import datetime
import matplotlib.pyplot as plt
```

```python
# NumPy arange() function
#Returns a NumPy array instead of a list, and supports floating-point steps. Stop value is exclusive.
#arr = np.arange(10)
#arr
arr1 = np.arange(10)
print(arr1)
arr2 = np.arange(5,50)
print(arr2)
arr3 = np.arange(2,20,0.2)
print(arr3)
```

```python
#Numpy linspace function
#create an array of evenly spaced numbers over a specified interval. Stop value is inclusive .
#Parameters: start, stop, num (number of points)
arr = np.linspace(1, 10, 5)  #include endpoint
print(arr)   #arr
arr1 = np.linspace(1,10,5, endpoint = False)   #exclude endpoint
print(arr1)  #arr1
arr2 = np.linspace(1,10,50,endpoint = False)
print(arr2)
```

```python
seed_value = int(datetime.datetime.now().timestamp())
print(seed_value)
random.seed(seed_value) # Seed the random number generator
random.seed()
print(random.random)
```

```python
data = np.random.rand(1000)
plt.hist(data, bins = 30 ,alpha =0.7, edgecolor ='black')
plt.title("Normal Distribution")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.savefig("Normal_Distribution.jpg")
plt.show()
```

```python
# randn function in NumPy

# Generate one random number from N(0,1)
x = np.random.randn()
print(x)
# Generate an array of 5 random numbers from N(0,1)
arr = np.random.randn(5)
print(arr)

# Generate a 3x3 matrix
matrix = np.random.randn(3, 3)
print(matrix)
```

```python
random_number =random.randint(1,100)  # Generate  random 
print(random_number)
```

```python
 x1 = np.random.normal(0, 0.8, 1000)
 x2 = np.random.normal(-2, 1, 1000)
 x3 = np.random.normal(3, 2, 1000)
 kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)
 plt.hist(x1, **kwargs)
 plt.hist(x2, **kwargs)
 plt.hist(x3, **kwargs);
```

```python
f = np.random.normal(loc=50, scale=10, size=5)  # mean=50, std=10
print("Random numbers from normal distribution:", f)
```
