```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from category_encoders.one_hot import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({'Size': np.random.choice(['XS','S','M','L','XL','XXL'],10),
                   'Brand': np.random.choice(['Nike','Puma', 'Adidas', 'Le Coq','Reebok'],10) })
df_train, df_test = train_test_split(df, test_size =0.5)
#print(df_train)
#print(df_test)
encoder = OneHotEncoder(use_cat_names= True, handle_unknown ='return Nan')
x_train = encoder.fit_transform(df_train)
x_test = encoder.fit_transform(df_test)
#print(x_train)
```

```python
colors = np.array(['Red', 'Green', 'Blue', 'Green', 'Red', 'Blue']).reshape(-1, 1)
ohe = OneHotEncoder()
encoded_colors = ohe.fit_transform(colors).toarray()

print("Original Data:\n", colors.flatten())
print("\nOne-Hot Encoded Array:\n", encoded_colors)
print("\nFeature Names:", ohe.get_feature_names_out(['Color']))
```

```python
colors = ['Red', 'Green', 'Blue', 'Green', 'Red', 'Blue']
le = LabelEncoder()
encoded_colors = le.fit_transform(colors)

print("Original Data:", colors)
print("Label Encoded Data:", encoded_colors)
print("Class Mapping:", dict(zip(le.classes_, range(len(le.classes_)))))
```
