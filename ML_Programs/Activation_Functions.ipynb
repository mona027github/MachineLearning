```python
# Sigmoid Activation function - Code
def sigmoid(x):
 return 1 / (1 + np.exp(-x))
x = np.linspace(-6,6,21)
print(x)
y = sigmoid(x)
plt.plot(x,y)
plt.grid()
plt.xlim(-6,6)
plt.xlabel('x')
plt.title('sigmoid(x)')
plt.savefig('C:/Users/Asus/ML_Programs/Datasets/sigmoid(x).jpg')
plt.show()
```

```python
# Logit (Logistic) - using logit() function
from scipy.special import logit
x = np.linspace(0, 1, 100)
y = logit(x)
print(y)
plt.plot(x, y)
plt.grid()
plt.ylim(-6, 6)
plt.xlabel('x')
plt.title('logit(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def logit(p):
  # """ Calculates the logit of a given probability p.p must be between 0 and 1 (exclusive)."""
  if not (0 < p < 1):
    raise ValueError("Probability p must be strictly between 0 and 1.")
  return np.log(p / (1 - p))

# Generate a range of probabilities
probabilities = np.linspace(0.01, 0.99, 100) # Exclude 0 and 1 to avoid division by zero or log(0)

# Calculate the logit for each probability
logit_values = [logit(p) for p in probabilities]

# Plot the logit function
plt.figure(figsize=(8, 6))
plt.plot(probabilities, logit_values, color='blue')
plt.title('Logit Function')
plt.xlabel('Probability (p)')
plt.ylabel('Logit (log(p/(1-p)))')
plt.grid(True)
plt.axvline(x=0.5, color='gray', linestyle='--', label='p = 0.5')
plt.axhline(y=0, color='gray', linestyle='--', label='Logit = 0')
plt.legend()
plt.show()
```
