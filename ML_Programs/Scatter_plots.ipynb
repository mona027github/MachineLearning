```python
import numpy as np
import matplotlib.pyplot as plt
X = np.random.randn(300,2)
plt.figure(figsize =(7.5,3.5))
plt.scatter(X[:,0], X[:,1], s=20)
plt.savefig('C:/Users/Asus/ML_Programs/scatter.jpg')
plt.show()
```

```python
# scatter options :
# s - size of point
# c - color of point
# cmap - colormap
# aplha - transparency ( 0 - 1)
#          1.0 - fully opaque
#          0.5 - semi-transparent
#          0.1 - very faint
```
