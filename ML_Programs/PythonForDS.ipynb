# Lists -  A list is simply an ordered collection.

```python
integer_list = [1, 2, 3]                                  # Simple list 
heterogeneous_list = ["string", 0.1, True]                # List with heterogeneous objects 
list_of_lists = [ integer_list, heterogeneous_list, [] ]  # List within list
list_length = len(integer_list)                           # Length of list   
list_sum    = sum(integer_list)                           # Sum of list
print(list_length)
print(list_sum)
```

```python
x = range(10)   # is the list [0, 1, ..., 9]
zero = x[0]     # equals 0, lists are 0-indexed
one = x[1]       # equals 1
nine = x[-1]   # equals 9, 'Pythonic' for last element
eight = x[-2]   # equals 8, 'Pythonic' for next-to-last element
x[0] = -1       # now x is [-1, 1, 2, 3, ..., 9]
```

```python
first_three   = x[:3]                     # [-1, 1, 2]              
three_to_end = x[3:]                      # [3, 4, ..., 9]            
one_to_four = x[1:5]                      # [1, 2, 3, 4]           
last_three = x[-3:]                       # [7, 8, 9]           
without_first_and_last = x[1:-1]          # [1, 2, ..., 8] 
copy_of_x = x[:]                          # [-1, 1, 2, ..., 9]              
```

```python
print( 1 in [1, 2, 3])# True
0 in [1, 2, 3]    # False
```

#  Generators and Iterators

```python
One way to create generators is with functions and the yield operator
def lazy_range(n): # """a lazy version of range"""
 i = 0
 while i < n:
   yield i
   i += 1
lrange = lazy_range(10)
print(lrange)
```
