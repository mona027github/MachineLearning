```python
#Numpy linspace function
#create an array of evenly spaced numbers over a specified interval. Stop value is inclusive .
#Parameters: start, stop, num (number of points)
arr = np.linspace(1, 10, 5)  #include endpoint
print(arr)   #arr
arr1 = np.linspace(1,10,5, endpoint = False)   #exclude endpoint
print(arr1)  #arr1
arr2 = np.linspace(1,10,50,endpoint = False)
print(arr2)
```
