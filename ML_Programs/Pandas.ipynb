```python
# Dataframe sample function
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 22, 35, 28]}
df = pd.DataFrame(data)
#df
sampled_rows = df.sample(n=2) #sample() func : Generate diff samples everytime
print(sampled_rows)
```

```python
sampled_rows = df.sample(n=3)
print("Selected 3 random rows:",sampled_rows)
sampled_fraction = df.sample(frac = 0.3)
print("Selected 30% of the rows randomly:\n",sampled_fraction)
sampled_with_replacement = df.sample(n=5, replace = True)
print("Sampled with replacement (allowing duplicates):")
print(sampled_with_replacement)

```

```python
# loc() function 

data={'Name':['Alice','Bob','Charlie','David', 'Eve'],
      'Age': [25, 32, 28, 22, 30],
      'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']}
df = pd.DataFrame(data) 
# access a single row
single_row = df.loc[2]
print(single_row,"\n")
 
# access rows 0, 3 and 4
row_list = df.loc[[0 ,3, 4]]
print(row_list)
print("\n")

# access a list of columns
column_list = df.loc[:,['Name','Age']]
print(column_list)
print("\n")

# access second row of 'Name' column
specific_value = df.loc[1, 'Name']
print(specific_value)



```
