```python
import numpy as np
import pandas as pd
```

```python
#Create an Array 
x = np.array([1,2,3,4])
print(x)
print("Element at index 2:", x[2])
print(type(x))
```

```python
# list
y = [1, 2, 3, 4]
print("First element:", y[0])
print(type(y))
```

```python
# Series in Pandas
data = [10, 20, 30, 40, 50]
my_series = pd.Series(data)
print(my_series)
print(my_series[2])
```

```python
a = [1, 2, 5]
a_series = pd.Series(a, index = ["x", "y", "z"])
print(a_series)
```

```python
 # create a dictionary with key value pairs
grades = {"Sem1" : 8.25, "Sem2" : 9.5, "Sem3" :7.75}
 # create a series from the dictionary
my_series = pd.Series(grades)
print(my_series)

first_year = pd.Series(grades, index =["Sem1", "Sem2"])
print(first_year)
```

```python
mat = np.matrix([[1, 2], [3, 4]])
print("Matrix:\n", mat)
print("Element at (2,1):", mat[1, 0])
```
