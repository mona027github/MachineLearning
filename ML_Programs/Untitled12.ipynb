# Clustering Techniques
 ● Clustering techniques are ML approaches that attempt to find similar patterns and relations between data samples in the dataset, and then cluster these samples into different groups so that every group or cluster of data samples has some correlation based on the actual attributes or characteristics.
 ● Clustering approaches are fully unsupervised as they seek to cluster data by looking at data features without prior training, guidance, or knowledge of data features, relations, and associations.
 ● It is the task of identifying similar instances and assigning them to clusters, or groups of similar instances.
 Clustering is a great tool for data analysis, customer segmentation, recommender systems, search engines, image segmentation, semi-supervised learning, dimensionality reduction etc.

 Types of Clustering Approaches
 - Centroid-based approaches, such as K-medoids and K-means
 - Hierarchical clustering approaches, such as divisive and agglomerative
 - Distribution-based clustering approaches, such as Gaussian mixture models
 - Density-based techniques such as optics dbscan (DBSCAN)
