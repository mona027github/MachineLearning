```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('mnist_train.csv')
test = pd.read_csv('mnist_test.csv')

X_train = train.iloc[:, 1:]
y_train = train.iloc[:, 0]
X_test = test.iloc[:, 1:]
y_test = test.iloc[:, 0]

clf = DecisionTreeClassifier(criterion='entropy', max_depth=10, random_state=42)
clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)
print(" Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))

plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='plasma')
plt.title("Decision Tree - MNIST Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

```

```python
plt.figure(figsize=(20, 10))
plot_tree(
    clf,
    max_depth=3,  # show only top 3 levels for readability
    filled=True,
    fontsize=8,
    feature_names=[f'pixel_{i}' for i in range(X_train.shape[1])],
    class_names=[str(i) for i in range(10)]
)
plt.title("Top 3 Levels of Decision Tree (MNIST)")
plt.show()

```
