# Data Transformation: It is a process to transform or consolidate the data in a form suitable for machine learning algorithms.
# Major techniques of data transformation are :
 • Normalization
 • Aggregation 
 • Generalization 
 • Feature construction
 
 **Normalization**: It is the technique of mapping the numerical feature values of any range into a specific smaller range i.e. 0 to 1                       or -1 to 1.
                    Popular methods of Normalization are:
                    1. Min-Max method
                    2. Mean normalization
                    3. Z score method 
                    4. Decimal scaling method



```python
 # Min-Max method
  """ X' = (X - Xmin)/(Xmax -Xmin)
   where, 
    X' is mapped value
    X is data value to be mapped in specific range
    Xmin and Xmax is minimum and maximum value of feature vector corresponding to X """
          
```

```python
import numpy as np
def min_max(X,Xmin,Xmax):
 return (X - Xmin)/(Xmax -Xmin)

# For a single value
print(min_max(30, 10, 50))  

X = np.array([2, 47, 90, 18, 5])
print("\n",.4f,min_max(X, X.min(), X.max()))

```
