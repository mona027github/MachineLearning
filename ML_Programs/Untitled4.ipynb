```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['row1', 'row2', 'row3'])
    # Select a single cell using labels
print(df.loc['row2', 'B'])

    # Select multiple rows and columns using labels
print(df.loc[['row1', 'row3'], ['A', 'B']])

    # Select rows based on a condition
print(df.loc[df['A'] > 1])
```

```python
    import pandas as pd

    df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['row1', 'row2', 'row3'])
    # Select a single cell using integer positions
    print(df.iloc[1, 1])

    # Select multiple rows and columns using integer positions
    print(df.iloc[[0, 2], [0, 1]])

    # Select a range of rows (exclusive of the end)
    print(df.iloc[0:2])
```
