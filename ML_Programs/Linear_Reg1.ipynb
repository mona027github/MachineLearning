```python
# Linear Regression example 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]).reshape(-1, 1)
Y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
summary = {
    "X_mean": np.mean(X),
    "Y_mean": np.mean(Y),
    "X_std": np.std(X, ddof=1),
    "Y_std": np.std(Y, ddof=1),
    "Correlation": np.corrcoef(X.flatten(), Y)[0, 1]
}

print("Summary Statistics")
for k, v in summary.items():
    print(f"{k}: {v:.4f}")

model = LinearRegression()
model.fit(X, Y)

slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, Y)

print("\n Regression Results")
print(f"Intercept: {intercept:.4f}")
print(f"Slope: {slope:.4f}")
print(f"R2: {r_squared:.4f}")

plt.scatter(X, Y, color="blue", label="Data Points")
plt.plot(X, model.predict(X), color="red", label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression: Y vs X")
plt.legend()
plt.grid(True)
plt.show()

# to comment out mulitple lines use c: ctrl+/

#  1) The data set "behaves like" a linear curve with some scatter;      
#  2) There is no justification for a more complicated model (e.g., quadratic);
#  3) There are no outliers;
#  4) The vertical spread of the data appears to be of equal height irrespective of the X-value; this indicates that the data are
#     equally-precise throughout and so a "regular" (that is, equi-weighted) fit is appropriate.
```
