```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data   # Features: sepal length, sepal width, petal length, petal width
y = iris.target # Target labels
feature_names = iris.feature_names

# 2. Standardize the data (zero mean, unit variance)
X_meaned = X - np.mean(X, axis=0)

# 3. Compute the covariance matrix
cov_matrix = np.cov(X_meaned, rowvar=False)

# 4. Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# 5. Sort eigenvalues and eigenvectors in descending order
sorted_index = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_index]
eigenvectors = eigenvectors[:, sorted_index]

# 6. Select top k eigenvectors (say k=2 for 2D visualization)
k = 2
eigenvectors_subset = eigenvectors[:, 0:k]

# 7. Transform the data
X_reduced = np.dot(X_meaned, eigenvectors_subset)

# 8. Plot the transformed data
plt.figure(figsize=(8,6))
for target, color, label in zip([0,1,2], ['r', 'g', 'b'], iris.target_names):
    plt.scatter(X_reduced[y == target, 0], 
                X_reduced[y == target, 1], 
                color=color, label=label)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Iris Dataset (manual implementation)")
plt.legend()
plt.grid(True)
plt.show()
```
