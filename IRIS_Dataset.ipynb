```python
# Histograms of the Features
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x_index = 3
colors = ['blue', 'red', 'green']
for label , color in zip(range(len(iris.target_names)), colors):
    ax.hist(
        iris.data[iris.target == label, x_index],
        label = iris.target_names[label],
        color = color)
    ax.set_xlabel(iris.feature_names[x_index])
    ax.legend(loc = 'upper right')
fig.show()
```

```python
# Scatter plot
fig, ax = plt.subplots()
x_index = 3 
y_index = 0
df = pd.DataFrame(iris.data, columns = iris.feature_names)
#df
colors = ['blue', 'red', 'green']
for label, color in zip(range(len(iris.target_names)), colors):
    df1 = df.iloc[iris.target == label ,3]
    df2 = df.iloc[iris.target == label ,0]
    valx = df1.values
    valy = df2.values
    ax.scatter(valx, valy, label = iris.target_names[label], c=color)
    ax.set_xlabel(iris.feature_names[3])
    ax.set_ylabel(iris.target_names[0])
    ax.legend(loc = 'upper left')
plt.show()
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(    
    scatter.legend_elements()[0], 
    iris.target_names, loc="lower right", title="Classes")
```
