```python
import pandas as pd
import math
data =  [9, 21, 29, 28, 4, 21, 8, 24, 26]
#  Step1: sorted the data
print("Given data : ",data)
data.sort()
print("Sorted data : ", data)
df = pd.DataFrame(data)
# Step2 :  Bin size calculation
bin_size = math.ceil((max(data) - min(data))/ len(data))
print("Bin Size :", bin_size)
#  Step 3 : Bin partitioning (equi-size)
bins = [data[i:i+bin_size] for i in range(0, len(data), bin_size)]
for idx, b in enumerate(bins, start=1):
    f"Bin {idx}: {b}"
# Step 4 : data smoothening
# Using mean value
bin_means = [sum(b)/len(b) for b in bins]
for idx, (b, mean) in enumerate(zip(bins, bin_means), start=1):
    print(f"Bin {idx}: {b}, Mean = {mean:.0f}")
    
# binned_with_means.append([bin_means] * len(b))

# # Flatten the list of lists
# smoothed_data = [val for sublist in binned_with_means for val in sublist]

# # Print results
# print("Original (sorted):", data)
# print("Binned with means:", smoothed_data)


```
